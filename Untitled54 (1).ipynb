```python
#4
!pip install -q faiss-cpu spacy


import faiss
import numpy as np
import spacy

# ✅ Load spaCy model with word vectors
nlp = spacy.load("en_core_web_md")

# ✅ Example words
words = ["king", "queen", "man", "woman", "apple", "orange"]

# ✅ Convert words → vectors
vectors = np.array([nlp(w).vector for w in words]).astype('float32')

# ✅ Create FAISS index (dimension = vector size)
dim = vectors.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(vectors)

# ✅ Query word
query_word = "royal"
query_vec = np.array([nlp(query_word).vector], dtype='float32')

# ✅ Search top 3 similar words
distances, indices = index.search(query_vec, k=3)

# ✅ Print results
print(f"\nQuery Word: '{query_word}'")
print("Top 3 Nearest Words:")
for i, idx in enumerate(indices[0]):
    print(f"{i+1}. {words[idx]}  (distance: {distances[0][i]:.4f})")
```

```python
#code 12
import random

class Agent:
    def __init__(self, name):
       self.name = name
    def act(self):
       return random.choice(["cooperate","defect"])

a1,a2=Agent("A"),Agent("B")
for _ in range(5):
    m1,m2=a1.act(),a2.act()
    outcome = "both win" if m1==m2=="cooperate" else "conflict" if m1!=m2 else "both lose"
    print(f"{a1.name}:{m1} {a2.name}:{m2} -> {outcome}")
    def simulate(steps=20):
    actions = ['cooperate', 'defect']
    score = {'A': 0, 'B': 0}
    history = []
    for t in range(steps):
        a1, a2 = random.choice(actions), random.choice(actions)
        if (a1, a2) == ('cooperate', 'cooperate'):
            score['A'] += 3; score['B'] += 3
        elif (a1, a2) == ('cooperate', 'defect'):
            score['A'] += 0; score['B'] += 5
        elif (a1, a2) == ('defect', 'cooperate'):
            score['A'] += 5; score['B'] += 0
        else:
            score['A'] += 1; score['B'] += 1
        history.append((t, a1, a2, dict(score)))
    return history

if __name__ == '__main__':
    for h in simulate(20):
        print(h)
```

```python
!pip install transformers datasets peft accelerate bitsandbytes -q

from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from peft import get_peft_model, LoraConfig, TaskType
import torch, os, math

os.environ["WANDB_DISABLED"] = "true"

# 1️⃣ Create small local dataset
with open("data.txt", "w") as f:
    f.write("""Artificial Intelligence is transforming the world.
Machine Learning enables systems to learn automatically.
Generative AI creates new data, music, and art.
Natural Language Processing allows machines to understand human language.""")

# 2️⃣ Load dataset
data = load_dataset("text", data_files={"train": "data.txt", "test": "data.txt"})

# 3️⃣ Load tokenizer & model
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)

# 4️⃣ Apply LoRA
config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["c_attn"],
    lora_dropout=0.05,
    task_type=TaskType.CAUSAL_LM
)
model = get_peft_model(model, config)

# 5️⃣ Tokenize text with labels
def tokenize_fn(example):
    tokens = tokenizer(example["text"], truncation=True, padding="max_length", max_length=64)
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized = data.map(tokenize_fn, batched=True)

# 6️⃣ Training setup
training_args = TrainingArguments(
    output_dir="./output",
    per_device_train_batch_size=1,
    num_train_epochs=1,
    logging_steps=1,
    save_steps=5,
    report_to="none"
)

# 7️⃣ Train
trainer = Trainer(model=model, args=training_args, train_dataset=tokenized["train"])
trainer.train()

# 8️⃣ Generate text
input_text = "AI will"
inputs = tokenizer(input_text, return_tensors="pt")
outputs = model.generate(**inputs, max_new_tokens=40, temperature=0.8, do_sample=True)
print("\nGenerated Text:\n", tokenizer.decode(outputs[0], skip_special_tokens=True))

# 9️⃣ Evaluate perplexity
eval_results = trainer.evaluate(tokenized["test"])
print("\nPerplexity:", math.exp(eval_results["eval_loss"]))
```
